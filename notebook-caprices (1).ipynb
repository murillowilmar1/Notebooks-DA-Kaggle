### Ocean_proximity

```python
print('ocean_proximity:',  df['ocean_proximity'].unique(),  '\n')
df["ocean_proximity"].replace(["NEAR BAY","<1H OCEAN", "INLAND","NEAR OCEAN","ISLAND"], [0,1,2,3,4 ], inplace=True)
```

```python
df.isnull().sum()
```

###  total_bedrooms

```python
x = df.total_bedrooms
plt.hist(x)
plt.show
mean = df.total_bedrooms.mean()
median =df.total_bedrooms.median()
mode = df.total_bedrooms.mode()
Columns = ["mean","median","mode"]
rows = [mean, median, mode]
data_I = pd.DataFrame(rows, Columns)
data_I
```

```python
df["total_bedrooms"] = df["total_bedrooms"].replace(np.nan, df.total_bedrooms.median())
```

```python
df["housing_median_age"]=df["housing_median_age"].apply(int)
df["total_rooms"]=df["total_rooms"].apply(int)
df["total_bedrooms"]=df["total_bedrooms"].apply(int)
df["population"]=df["population"].apply(int)
df["households"]=df["households"].apply(int)
df["median_house_value"]=df["median_house_value"].apply(int)
```

# ML_MODEL - GradientBoostingRegressor

```python


from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split



```

```python
# Almacenar Variables seleccionadoas para el algoritmo 
Car_select =["housing_median_age","total_rooms","total_bedrooms","population","households","median_income","ocean_proximity"]
#Almacenar la variable objetivo 
obj_select = df.median_house_value

# Almacenar datos en variables pequeñas para manipular datos en el modelo 
X = df[Car_select]
y = obj_select
```

```python
X_train, X_test, y_test, y_train = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python
DG_model =GradientBoostingRegressor()
DG_model.fit(X, y)
```

```python
print(DG_model.predict(X.head()))
print(df.median_house_value.head())
```

```python
DG_model.score(X,y)
```
